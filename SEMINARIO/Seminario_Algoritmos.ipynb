Se implementa, y ejecuta, un algoritmo por fueza bruta que prueba todas las asignaciones posibles de horarios.

````python
# Imports
import random
from collections import Counter
from itertools import product
import threading
import time
````

````python
# Excepciones personalizadas
class TimeoutException(Exception):
    pass
````

````python
# Estructura de datos
horarios = [
    "V20",
    "S12",
    "S16",
    "S18",
    "S20",
    "D12",
    "D16",
    "D18",
    "D20",
    "L20"
]

idx_viernes = horarios.index("V20")
idx_lunes = horarios.index("L20")

categorias = ["A", "B", "C"]

# Audiencia base (millones)
audiencia_base = {
    ("A", "A"): 2.0,
    ("A", "B"): 1.3,
    ("A", "C"): 1.0,
    ("B", "A"): 1.3,
    ("B", "B"): 0.9,
    ("B", "C"): 0.75,
    ("C", "A"): 1.0,
    ("C", "B"): 0.75,
    ("C", "C"): 0.47
}

# Coeficientes por horario
coef_horario = {
    "V20": 0.4,
    "S12": 0.55,
    "S16": 0.70,
    "S18": 0.80,
    "S20": 1.00,
    "D12": 0.45,
    "D16": 0.75,
    "D18": 0.85,
    "D20": 1.00,
    "L20": 0.40
}

# Penalización por coincidencias
penalizacion = {
    0: 0.00,
    1: 0.25,
    2: 0.45,
    3: 0.60,
    4: 0.70,
    5: 0.75,
    6: 0.78,
    7: 0.80,
    8: 0.80
}
````

````python
# Funciones auxiliares
def audiencia_sin_penalizacion(partido, horario):
    return (audiencia_base[partido] * coef_horario[horario])

def calcular_audiencia(partidos, asignacion):
    conteo = Counter(asignacion)

    total = 0

    for partido, horario in zip(partidos, asignacion):

        base = audiencia_sin_penalizacion(partido, horario)

        coincidencias = conteo[horario] - 1

        reduccion = penalizacion[coincidencias]

        total += base * (1 - reduccion)

    return total

def cumple_restricciones(asignacion):
    # Verifica si la asignación cumple con las restricciones:
    tiene_viernes = horarios[idx_viernes] in asignacion # Al menos un partido el viernes
    tiene_lunes = horarios[idx_lunes] in asignacion # Al menos un partido el lunes
    return (tiene_viernes and tiene_lunes)
````

````python
# Algoritmo por fuerza bruta
def maximizar_audiencia_fuerza_bruta(partidos, stop_event):
    mejor_audiencia = -1
    mejor_asignacion = None
    
    # Genera por fuerza bruta todas las combinaciones posibles de horarios para los 10 partidos
    # Cada elemento en 'asignacion' es una tupla donde el índice es el partido y el valor es el horario.
    # Ejecutar este bloque de código iterará sobre los 10 mil millones de opciones. 
    combinaciones = product(horarios, repeat=10)
    #combinaciones = product(horarios, repeat=5)

    for asignacion in combinaciones:
        # Verificar si el hilo principal activó la bandera de parada
        if stop_event.is_set():
            raise TimeoutException()
            
        # Filtrar por restricciones
        if not cumple_restricciones(asignacion):
            continue
            
        # Calcular la audiencia total de esta combinación específica
        audiencia_actual = calcular_audiencia(partidos, asignacion)
        
        # Mantener el máximo global
        if audiencia_actual > mejor_audiencia:
            mejor_audiencia = audiencia_actual
            mejor_asignacion = asignacion
            
    return mejor_audiencia, mejor_asignacion

resultado_fb = {}

# Algoritmo por fuerza bruta con evento de parada
def maximizar_audiencia_fuerza_bruta_seguro(partidos, stop_event):
    try:
        mejor_audiencia_fb, mejor_asignacion_fb = maximizar_audiencia_fuerza_bruta(partidos, stop_event)
        resultado_fb["mejor_audiencia"] = mejor_audiencia_fb
        resultado_fb["mejor_asignacion"] = mejor_asignacion_fb
    except TimeoutException:
        resultado_fb["mejor_audiencia"] = None
        resultado_fb["mejor_asignacion"] = None
````

````python
# Jornada de descripción del problema
partidos_descripcion = [
    ('B', 'A'),
    ('B', 'A'),
    ('C', 'C'),
    ('B', 'A'),
    ('C', 'C'),
    ('B', 'C'),
    ('B', 'B'),
    ('B', 'B'),
    ('B', 'C'),
    ('B', 'B')
]
````

````python
print("Buscando la mejor combinación por fuerza bruta...")

# Inicializar la bandera de parada
evento_parada = threading.Event()

# Crear e iniciar el hilo pasando la bandera como argumento
hilo = threading.Thread(target = maximizar_audiencia_fuerza_bruta_seguro, args = (partidos_descripcion, evento_parada))
hilo.daemon = True

inicio = time.perf_counter()

hilo.start()

# Esperar el tiempo límite
hilo.join(timeout=1800) # Tiempo limite en segundos

if hilo.is_alive():
    fin = time.perf_counter()
    tiempo = fin - inicio
    print("Tiempo de espera terminado")    
    # Activar la bandera para que el hilo secundario muera de inmediato
    evento_parada.set()  
    # Espera un breve instante para confirmar el cierre limpio en consola
    hilo.join()
else:
    fin = time.perf_counter()
    tiempo = fin - inicio
    print(f"Mejor audiencia: {resultado_fb["mejor_audiencia"]} millones")

    print("Asignación óptima:")
    
    for i, horario in enumerate(resultado_fb["mejor_asignacion"]):
        print(f"Partido {i+1}: " f"{partidos[i]} -> {horario}")
        
print(f"Ejecución terminada en {tiempo} segundos")
````

**(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta**

Respuesta

Se propone utilizar un algoritmo de retroceso (backtracking) con poda (bound). Este algoritmo consta de los siguientes pasos:

1. Asignar partidos uno a uno.
2. Calcular audiencia parcial.
3. Estimar una cota superior de la audiencia máxima posible.
4. Si esa cota ya es peor que la mejor solución encontrada, se poda la rama.

Con esto conseguimos:
- Evitar explorar millones de combinaciones inútiles.
- Mantener la optimalidad.
- Explorar una fracción del espacio.

````python
# Algoritmo retroceso (backtracking) con poda (bound)
def maximizar_audiencia_retroceso(partidos, stop_event):
    n = len(partidos)
    
    # Máxima contribución teórica de cada partido
    maximos = [audiencia_base[p] for p in partidos]
    
    # Sumas acumuladas de máximos
    sufijos = [0] * (len(partidos) + 1)

    for i in range(len(partidos) - 1, -1, -1):
        sufijos[i] = sufijos[i + 1] + maximos[i]
    
    mejor_audiencia = -1
    mejor_asignacion = None

    def retroceso(indice, asignacion_actual, audiencia_parcial, stop_event):
        nonlocal mejor_audiencia
        nonlocal mejor_asignacion
        
        # Poda
        cota = (audiencia_parcial + sufijos[indice])

        if cota <= mejor_audiencia:
            return
        
        # Caso base
        if indice == n:
            # Filtrar por restricciones
            if not cumple_restricciones(asignacion_actual):
                return

            audiencia_real = calcular_audiencia(partidos, asignacion_actual)

            if audiencia_real > mejor_audiencia:
                mejor_audiencia = audiencia_real
                mejor_asignacion = asignacion_actual.copy()

            return

        # Verificar si el hilo principal activó la bandera de parada
        if stop_event.is_set():
            raise TimeoutException()
            
        # Expandir nodo
        partido = partidos[indice]

        for horario in horarios:
            contribucion = audiencia_sin_penalizacion(partido, horario)

            nueva_audiencia = (audiencia_parcial + contribucion)

            asignacion_actual.append(horario)

            retroceso(indice + 1, asignacion_actual, nueva_audiencia, stop_event)

            asignacion_actual.pop()

    retroceso(0, [], 0, stop_event)

    return mejor_audiencia, mejor_asignacion

resultado_rp = {}

# Algoritmo retroceso (backtracking) con poda (bound) y evento de parada
def maximizar_audiencia_retroceso_seguro(partidos, stop_event):
    try:
        mejor_audiencia_rp, mejor_asignacion_rp = maximizar_audiencia_retroceso(partidos, stop_event)
        resultado_rp["mejor_audiencia"] = mejor_audiencia_rp
        resultado_rp["mejor_asignacion"] = mejor_asignacion_rp
    except TimeoutException:
        resultado_rp["mejor_audiencia"] = None
        resultado_rp["mejor_asignacion"] = None
````

````python
print("Buscando la mejor combinación por retroceso (backtracking) con poda (bound)...")

# Inicializar la bandera de parada
evento_parada = threading.Event()

# Crear e iniciar el hilo pasando la bandera como argumento
hilo = threading.Thread(target = maximizar_audiencia_retroceso_seguro, args = (partidos_descripcion, evento_parada))
hilo.daemon = True

inicio = time.perf_counter()

hilo.start()

# Esperar el tiempo límite
hilo.join(timeout=3600) # Tiempo limite en segundos

if hilo.is_alive():
    fin = time.perf_counter()
    tiempo = fin - inicio
    print("Tiempo de espera terminado")    
    # Activar la bandera para que el hilo secundario muera de inmediato
    evento_parada.set()  
    # Espera un breve instante para confirmar el cierre limpio en consola
    hilo.join()
else:
    fin = time.perf_counter()
    tiempo = fin - inicio
    print(f"Mejor audiencia: {resultado_rp["mejor_audiencia"]} millones")

    print("Asignación óptima:")
    
    for i, horario in enumerate(resultado_rp["mejor_asignacion"]):
        print(f"Partido {i+1}: " f"{partidos[i]} -> {horario}")
        
print(f"Ejecución terminada en {tiempo} segundos")
````

Encontramos que esta primera versión del algoritmo toma aproximadamente entre 25 a 40 minutos de tiempo de ejecución, dependiendo de la jornada de partidos, en el equipo de prueba.

Por esto motivo procedemos a realizar las siguientes mejoras:
- Precalcular audiencias
- Ordenar partidos
- Explorar mejores horarios primero
- Reemplazar Counter por lista

````python
# Función auxiliar
def obtener_penalizacion(conteo):
    idx = max(0, conteo - 1)
    return penalizacion.get(idx, penalizacion[max(penalizacion)])

# Algoritmo retroceso (backtracking) con poda (bound) mejorado
def maximizar_audiencia_retroceso_mejorado(partidos, stop_event):
    n = len(partidos)
    m = len(horarios)
    
    # Precalcular audiencias
    audiencias = [[audiencia_base[p] * coef_horario[h] for h in horarios] for p in partidos]
    
    # Ordenar partidos
    orden = sorted(range(n), key = lambda i: max(audiencias[i]), reverse = True)

    partidos = [partidos[i] for i in orden]
    audiencias = [audiencias[i] for i in orden]

    # Sumas acumuladas de máximos
    sufijos = [0] * (n + 1)

    for i in range(n - 1, -1, -1):
        sufijos[i] = sufijos[i + 1] + max(audiencias[i])

    mejor_audiencia = -1
    mejor_asignacion = None

    asignacion = [-1] * n

    conteo = [0] * m

    partidos_horario = [[] for _ in range(m)]

    contribucion = [0.0] * n

    audiencia_actual = 0.0
    
    def retroceso_mejorado(indice, tiene_viernes, tiene_lunes):
        nonlocal mejor_audiencia
        nonlocal mejor_asignacion
        nonlocal audiencia_actual
        
        # Poda
        cota = (audiencia_actual + sufijos[indice])
        
        if cota <= mejor_audiencia:
            return
            
        # Caso base
        if indice == n:
            if not (tiene_viernes and tiene_lunes):
                return
                
            if audiencia_actual > mejor_audiencia:
                mejor_audiencia = audiencia_actual
                mejor_asignacion = asignacion.copy()
                
            return

        # Verificar si el hilo principal activó la bandera de parada
        if stop_event.is_set():
            raise TimeoutException()
            
        # Probar horarios
        orden_horarios = sorted(range(m), key = lambda h: audiencias[indice][h], reverse = True)

        for h in orden_horarios:
            # Guardar estado
            audiencia_anterior = audiencia_actual

            afectados = partidos_horario[h].copy()

            contribuciones_antiguas = [contribucion[p] for p in afectados]
            
            # Insertar partido
            partidos_horario[h].append(indice)

            conteo[h] += 1

            asignacion[indice] = h
            
            penal = obtener_penalizacion(conteo[h])

            # Recalcular todos los partidos del horario
            for p in afectados:
                audiencia_actual -= contribucion[p]             
                contribucion[p] = (audiencias[p][h] * (1 - penal))
                audiencia_actual += contribucion[p]

            contribucion[indice] = (audiencias[indice][h] * (1 - penal))

            audiencia_actual += contribucion[indice]

            retroceso_mejorado((indice + 1), (tiene_viernes or (h == idx_viernes)), (tiene_lunes or (h == idx_lunes)))
            
            # Deshacer
            audiencia_actual = audiencia_anterior

            partidos_horario[h].pop()

            conteo[h] -= 1

            contribucion[indice] = 0

            for p, c in zip(afectados, contribuciones_antiguas):
                contribucion[p] = c

    retroceso_mejorado(0, False, False)
    
    if mejor_asignacion is None:
        return None, None
    
    # Recuperar orden original
    resultado = [None] * n

    for pos, original in enumerate(orden):
        resultado[original] = horarios[mejor_asignacion[pos]]

    return mejor_audiencia, resultado

resultado_rpm = {}
    
# Algoritmo retroceso (backtracking) con poda (bound) y evento de parada mejorado
def maximizar_audiencia_retroceso_mejorado_seguro(partidos, stop_event):
    try:
        mejor_audiencia_rpm, mejor_asignacion_rpm = maximizar_audiencia_retroceso_mejorado(partidos, stop_event)
        resultado_rpm["mejor_audiencia"] = mejor_audiencia_rpm
        resultado_rpm["mejor_asignacion"] = mejor_asignacion_rpm
    except TimeoutException:
        resultado_rpm["mejor_audiencia"] = None
        resultado_rpm["mejor_asignacion"] = None
````

````python
print("Buscando la mejor combinación por retroceso (backtracking) con poda (bound) mejorado...")

# Inicializar la bandera de parada
evento_parada = threading.Event()

# Crear e iniciar el hilo pasando la bandera como argumento
hilo = threading.Thread(target = maximizar_audiencia_retroceso_mejorado_seguro, args = (partidos_descripcion, evento_parada))
hilo.daemon = True

inicio = time.perf_counter()

hilo.start()

# Esperar el tiempo límite
hilo.join(timeout=3600) # Tiempo limite en segundos

if hilo.is_alive():
    fin = time.perf_counter()
    tiempo = fin - inicio
    print("Tiempo de espera terminado")    
    # Activar la bandera para que el hilo secundario muera de inmediato
    evento_parada.set()  
    # Espera un breve instante para confirmar el cierre limpio en consola
    hilo.join()
else:
    fin = time.perf_counter()
    tiempo = fin - inicio
    print(f"Mejor audiencia: {resultado_rpm["mejor_audiencia"]} millones")

    print("Asignación óptima:")
    
    for i, horario in enumerate(resultado_rpm["mejor_asignacion"]):
        print(f"Partido {i+1}: " f"{partidos[i]} -> {horario}")
        
print(f"Ejecución terminada en {tiempo} segundos")
````
